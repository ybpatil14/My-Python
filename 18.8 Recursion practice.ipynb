### Main function - Recursive

```python
def power_finder(my_list, final_op = None):
    
#     Initialization
    if final_op == None:
        final_op = my_list[0]
        my_list = my_list[1:]
    
    print(f'final element : {final_op}')
    print(f'Current list we are working on : {my_list}')
    print()
    
#     Break condition
    if len(my_list) == 0:
        return final_op
    
#     Processing condition
    else:
        final_op = final_op ** my_list[0]
        my_list = my_list[1:]
        return power_finder(my_list, final_op)
```

```python
def helper_function(list_):
    if not list_[0] in (0,1):
        return list_
    else:
        return helper_function(list_[1:])


def power_finder_arb(*my_list, final_op = None):    
#     Initialization
    if final_op == None:
        my_list = helper_function(my_list)
        final_op = my_list[0]
        my_list = my_list[1:]
    print(f'final element : {final_op}')
    print(f'Current list we are working on : {my_list}')
    print()
#     Break condition
    if len(my_list) == 0:
        return final_op
#     Processing condition
    else:
        final_op = final_op ** my_list[0]
        my_list = my_list[1:]
        return power_finder(my_list, final_op)
```

```python
# Gayatri22:15
def ascii_range(character, start_point, end_point):
    """This function will find the ascii value of given character.It takes tree inputs
    character:any character
    start_point:int
    end_point:int"""
    
    print(start_point, end_point)
    
    if chr(start_point)==character:
        return print(f"Ascii value of {character} is {start_point} in range{(start_point//100)*100,end_point}")
    if start_point==end_point:
        return ascii_range(character,start_point+1,end_point+100)
    else:
        start_point+=1
        return ascii_range(character,start_point,end_point)
```

```python
def asci_finder(element,start_ = None , end_ = None, batch_size = 500):
    if start_ == None and end_ == None:
        start_ = 0
        end_ = 100
    
#     print(start_, end_)
    for final in range(start_, end_):
#         print(f'ascii : {final}')
#         print(f'characters : {chr(final)}')

        if final > 1114111:
            print('Not Available !')
            break
        elif chr(final) == element:
            return final
        
    else:
        start_, end_ = end_, start_ + batch_size + end_
#         start_ += batch_size
        
        return asci_finder(element, start_, end_)
```
