### Recommended method

##### Using a context-manager

- Context manager is an approach that closes the file automatically for us without mentioning explicitly

```python
with open(file_path, 'r') as file_obj:
    data = file_obj.read()
```

```python
print(data)
```

```python
with open(file_path, 'r') as file_obj1:
    data_1 = file_obj1.readlines()
```

```python
print(data_1)
```

```python
len(data_1)
```

```python
with open(file_path, 'r') as file_obj2:
    data_2 = file_obj2.readline()
```

```python
data_2
```

```python
final = []
with open(file_path, 'r') as file_obj3:
    file_obj3.seek(10) # change the starting point
    
    print(file_obj3.readline())
    print(file_obj3.tell()) # find out the current point
```

- Read mode is used when we want to just work with or read the data
- Write mode is only/usually used when we want to generate new files
- Append mode is used when we want to add additional data to the already existing files

```python
# Wrong way to add the statement to the end of the already existing data
statement = 'I am a guardian of the galaxy !'

with open(file_path, 'w') as f:
    f.write(statement)
```

```python
# Wrong way to add the statement to the end of the already existing data
statement = 'I am a guardian of the galaxy !'

with open(file_path, 'w') as f:
    f.seek(20) # changing the starting reference point
    f.write(statement)
```

```python
# Right way to add the statement to the end of the already existing data
statement = 'I am a guardian of the galaxy !'

with open(file_path, 'a') as f:
    f.write(statement)
```
