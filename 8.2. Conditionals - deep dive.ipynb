### Bitwise operators

and, or, not

- condition1 and condition2 > both the consitions should be satisfied
- condition1 or condition2 > either condition1 or condition2 should be satisfied

- not condition1 > if condition1 is not satisfied
- not (condition1 and condition2) > if both the conditions dont satisfy at once
- not (condition1 or condition2) > if none of the conditions satisfy even individually

- and > &
- or > |
- not > !

```python
word = 'anime'
vowels = 'aeiouAEIOU'
if (word[0].islower() & vowels.__contains__(word[0])):
    print(1)
else:
    print(0)
```

```python
word = 'anime'
vowels = 'aeiouAEIOU'
if (word[0].islower() and vowels.__contains__(word[0])):
    print(1)
else:
    print(0)
```
